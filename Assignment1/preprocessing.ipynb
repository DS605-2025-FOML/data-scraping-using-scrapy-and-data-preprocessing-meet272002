```python
import pandas as pd
```

```python
dataframe = pd.read_csv('Assignment1.csv')
```

```python
dataframe
```

```python
dataframe = dataframe[['title','rating','price','availability']]
```

```python
dataframe.rename(columns={'title':'Title','rating':'Rating','price':'Price','availability':'Availability'},inplace=True)
```

```python
dataframe['Rating'].unique()
```

```python
dataframe.describe()
```

```python
dataframe['Price'] = dataframe['Price'].str.replace('£','',regex=False).astype(float)
```

```python
dataframe
```

```python
Rating = {'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
```

```python
dataframe['Rating'] = dataframe['Rating'].apply(lambda x: Rating[x])
```
